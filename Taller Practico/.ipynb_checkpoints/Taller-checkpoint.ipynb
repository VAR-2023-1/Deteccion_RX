```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

ruta="logo_un.jpeg"
img_bgr = cv2.imread(ruta, cv2.IMREAD_COLOR)
img_gray = cv2.imread(ruta, cv2.IMREAD_GRAYSCALE)
#Corregimos la lectura de BGR a RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Crea una figura con un tamao personalizado
fig, ax = plt.subplots(figsize=(8, 8))
# Muestra la imagen en la figura
ax.set_title('Imagen Original')
ax.imshow(img_rgb)


# Conversin a cada uno de los canales
img_rgb  = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_xyz  = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2XYZ)
img_lab  = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2LAB)
img_hsv  = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

# Discriminacin de cada uno de los canales de separacin
## Canal RGB
img_R = img_rgb[:, :, 0]
img_G = img_rgb[:, :, 1]
img_B = img_rgb[:, :, 2]
## Canal CMY
img_C = 255 - img_rgb[:,:,0]
img_M = 255 - img_rgb[:,:,1]
img_Y = 255 - img_rgb[:,:,2]

#Se obtienen los canales YIQ mediante una transformacin lineal
img_Y = 0.299*img_R + 0.587*img_G + 0.114*img_B

# Crea una figura con un tamao personalizado
fig, ax = plt.subplots(figsize=(8, 8))
# Muestra la imagen en la figura
ax.set_title('Imagen Original')
ax.imshow(img_rgb)
```

```python
from plantcv import plantcv as pcv
s_binario = pcv.threshold.binary(gray_img=img_Y,
                                 threshold=140,
                                 max_value=255,
                                 object_type='light')
s_mblur = pcv.median_blur(s_binario, ksize=40)

plt.imshow(s_binario,cmap='gray')
```

```python
# Crear una imagen de 3 canales RGB con el mismo tamao que la imagen original
imagen_rgb = np.zeros((s_mblur.shape[0], s_mblur.shape[1], 3), dtype=np.uint8)

# Copiar el canal de la imagen de 2 canales a los canales R y G de la imagen RGB
imagen_rgb[:,:,0] = s_mblur
imagen_rgb[:,:,1] = s_mblur

# Asignar un valor constante al canal B de la imagen RGB
imagen_rgb[:,:,2] = s_mblur# Puedes ajustar el valor segn tus necesidades
plt.imshow(imagen_rgb)
img_shape = np.shape(imagen_rgb)
img_datatype = imagen_rgb.dtype
print(img_shape)
print(img_datatype)
img_shape = np.shape(img_rgb)
img_datatype = img_rgb.dtype
print(img_shape)
print(img_datatype)
```

```python
sa_img = pcv.logical_and(img_rgb,imagen_rgb)

#Generar un subplot 1x3 para mostrar la imagen de entrada, la mscara y la operacin AND
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(15, 7))
ax1.set_title('Imagen de entrada')
ax2.set_title('Mscara')
ax3.set_title('Operacin lgica')
ax1.imshow(img_rgb, cmap='gray')
ax2.imshow(imagen_rgb, cmap='gray')
ax3.imshow(sa_img)
```

```python
resultado = cv2.cvtColor(sa_img, cv2.COLOR_RGB2GRAY)
plt.imshow(resultado,cmap='gray')
```
